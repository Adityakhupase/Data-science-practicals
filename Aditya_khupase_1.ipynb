```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import metrics

```

```python
df = pd.read_csv("/income.csv")
df
```

```python
sns.scatterplot(data=df[['Age','Income($)']],x='Age',y='Income($)')
```

```python
from sklearn.cluster import KMeans
KM_Model = KMeans(n_clusters = 3)
cluster_predict_old = KM_Model.fit_predict(df[['Age','Income($)']])
cluster_predict_old
```

```python
df['old_cluster'] = cluster_predict_old
df
```

```python
df0 = df[df['old_cluster']==0]
df1 = df[df['old_cluster']==1]
df2 = df[df['old_cluster']==2]
df0
```
